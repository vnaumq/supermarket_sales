```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

path_to_file = r'D:\university\VSCODE\project\Supermarket_sales\supermarket_sales.csv'
```

```python
df = pd.read_csv(path_to_file)
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'])
df.dtypes
```

```python
df.columns
```

```python
def rename_columns(name):
    new_name = name.lower().replace(' ','_')
    return new_name
```

```python
df.rename(columns= rename_columns,inplace=True)
```

```python
df
```

```python
df1 = df[['city','gender','total','date','payment']]
df1['month'] = pd.DatetimeIndex(df1['date']).month_name()
df1['weekday'] = pd.DatetimeIndex(df1['date']).day_name()
df1['weekday_number'] = pd.DatetimeIndex(df1['date']).weekday
df1
# Приведение даты в месяц и в день недели
```

```python
gender_sum = df1.groupby(['gender'],as_index=False).agg({'total':'sum'}).rename(columns={'total':'sum'})
```

```python
gender_sum
```

```python
city_sum = df1.groupby(['city'],as_index=False).agg({'total':'sum'}).rename(columns={'total':'sum'})
```

```python
city_sum.sort_values('sum',ascending=False)
```

```python
sns.histplot(data= df1, x= 'total',)
```

```python
sns.histplot(df1, x= 'total',  hue="gender",multiple='stack',fill = False)

```

```python
df1.groupby('payment',as_index=False).agg({'gender':'count', 'total' : 'sum'}).rename(columns = {'gender' : 'count', 'total' : 'sum'}).sort_values('sum',ascending = False)
```

```python
rating_weekday = df1.groupby(['month','weekday','weekday_number'] , as_index= False).agg({'total' : 'sum'}).rename(columns = {'total' : 'revenue'}).sort_values(['month', 'weekday_number'])
rating_weekday
```

```python
sns.barplot(rating_weekday,x = 'weekday', y = 'revenue', hue = 'month')
```
