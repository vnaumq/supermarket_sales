```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

path_to_file = r'D:\university\VSCODE\project\Supermarket_sales\supermarket_sales.csv'
```

```python
df = pd.read_csv(path_to_file)
df
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'])
df.dtypes
```

```python
df.columns
```

```python
def rename_columns(name):
    new_name = name.lower().replace(' ','_')
    return new_name
```

```python
df.rename(columns= rename_columns,inplace=True)
```

```python
df
```

```python
df1 = df[['city','gender','total','date','payment','time','product_line']]
df1['month'] = pd.DatetimeIndex(df1['date']).month_name()
df1['weekday'] = pd.DatetimeIndex(df1['date']).day_name()
df1['weekday_number'] = pd.DatetimeIndex(df1['date']).weekday
df1['hour'] = pd.DatetimeIndex(df1['time']).hour
df1['week'] = df1['date'].dt.strftime('%U').astype(int)

df1
# Приведение даты в месяц и в день недели
```

```python
count_week = df1.week.value_counts().reset_index().sort_values('week').reset_index().drop('index',axis=1)
count_week
```

```python
gender_sum = df1.groupby(['gender'],as_index=False).agg({'total':'sum'}).rename(columns={'total':'sum'})
```

```python
gender_sum
```

```python
city_sum = df1.groupby(['city'],as_index=False).agg({'total':'sum'}).rename(columns={'total':'sum'})
```

```python
df1.groupby('payment',as_index=False).agg({'gender':'count', 'total' : 'sum'}).rename(columns = {'gender' : 'count', 'total' : 'sum'}).sort_values('sum',ascending = False)
```

```python
rating_weekday = df1.groupby(['month','weekday','weekday_number'] , as_index= False).agg({'total' : 'sum'}).rename(columns = {'total' : 'revenue'}).sort_values(['month', 'weekday_number'])
rating_weekday
```

```python
sns.barplot(rating_weekday,x = 'weekday', y = 'revenue', hue = 'month')
```

```python
revenue_hour = df1.groupby(['month','hour'],as_index=False).agg({'total' : 'sum'}).rename(columns={'total':'revenue'})
revenue_hour
```

```python
sns.lineplot(revenue_hour,x='hour',y='revenue',hue='month',style="month")
```

```python
revenue_week = df1.groupby('week').agg({'total':'sum'}).reset_index().rename(columns={'total':'revenue'})
revenue_week
```

```python
sns.lineplot(revenue_week,x='week',y='revenue',marker='o',markerfacecolor = 'black')
```

```python
product_revenue = df1.groupby('product_line',as_index=False).agg({'total' : 'sum'}).rename(columns={'total':'revenue'}).sort_values('revenue',ascending=False)
product_revenue.set_index('product_line',inplace=True)
product_revenue
```

```python
product_revenue.plot.pie(y='revenue',figsize=(7,7))
```

```python
fig, ax = plt.subplots()
ax.pie(product_revenue.revenue, labels=product_revenue.index, autopct='%1.1f%%')
```

```python
explode = (0.07, 0.01, 0.01, 0.01, 0.01, 0.01)
fig, ax = plt.subplots()
ax.pie(product_revenue.revenue, labels=product_revenue.index, autopct='%1.1f%%', explode=explode,startangle=90)
```

```python
product_line_week = df1.groupby(['week','product_line']).agg({'total':'sum'}).reset_index().rename(columns={'total':'revenue'})
product_line_week
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(product_line_week,x='week',y='revenue',hue='product_line',style="product_line",)
```

```python

sns.relplot(
    data=product_line_week, x="week", y="revenue",
    col="product_line", hue="product_line", style="product_line",
    kind="line",marker='o',col_wrap=3,sizes=(10, 10)
)
```
